```python
def getPedido(pedido_id):
    pedido_id = ObjectId(pedido_id)
    pedido = collection.find_one({'_id': ObjectId(pedido_id)})
    return pedido
```

```python
def postPedido(pedido):
    res = collection.insert_one(pedido)
    return res.inserted_id
```

```python
def putPedido(pedido_id, update_pedido):
    pedido_id = {"_id": ObjectId(pedido_id)}
    update_pedido = {"$set": update_pedido}
    res = collection.update_one(pedido_id, update_pedido)
    return res.modified_count
```

```python
def deletePedido(pedido_id):
    pedido_id = {"_id": ObjectId(pedido_id)}
    res = collection.delete_one(pedido_id)
    return res.deleted_count
```

```python
import pandas as pd
```

```python
def postMany(documents):
    res = collection.insert_many(documents)
    return res.inserted_ids
```

```python
COLLECTION = "itensdisponiveis"
```

```python
collection = db[COLLECTION]
```

```python
%%time

pedidos = getPedidos()
pedidos
```

```python
fixos = pd.read_csv("../Fixos.csv", sep=";", encoding='latin') \
.drop(["N°", "Valor Unitário (R$)", "Quantidade Anual ", "Valor Anual "], axis=1) \
.rename({"Descrição ": "descricao", "Unid.": "unidade", "Fabricante Referência": "fabReferencia"}, axis=1)

fixos['unidade'] = fixos['unidade'].map({"Und.": "unid.", "metro": "m", "Kg": "kg"})
fixos['categoria'] = "Fixo"

fixos
```

```python
fixos['unidade'].value_counts()
```

```python
dicts = []
for index, row in fixos.iterrows():
    curr = {}
    for col in row.index:
        curr[col] = row[col]
        
    dicts.append(curr)
    
dicts
```

```python
pedido_id = postMany(dicts)
pedido_id
```

```python
sobdemanda = pd.read_csv("../SobDemanda.csv", sep=";", encoding='latin') \
.drop(["N°", "Valor Unitário (R$)", "Quantidade Anual Revisada", "Valor Anual Revisado"], axis=1) \
.rename({"Descrição Revisada": "descricao", "Unid.": "unidade", "Fabricante Referência": "fabReferencia"}, axis=1)

sobdemanda['unidade'] = sobdemanda['unidade'].map({"Und.": "unid.", "metro": "m", "Kg": "kg", "m²": "m²"})
sobdemanda['descricao'] = sobdemanda['descricao'].apply(lambda x: x.strip())

sobdemanda['categoria'] = "Sob Demanda"

sobdemanda
```

```python
sobdemanda['unidade'].value_counts()
```

```python
dicts = []
for index, row in sobdemanda.iterrows():
    curr = {}
    for col in row.index:
        curr[col] = row[col]
        
    dicts.append(curr)
    
dicts
```

```python
pedido_id = postMany(dicts)
pedido_id
```
